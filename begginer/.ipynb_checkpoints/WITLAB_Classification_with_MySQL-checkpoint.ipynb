```python
import MySQLdb
import os
import numpy as np
import csv

# MySQL DB 연결
db = MySQLdb.connect('210.102.142.13',"root", "witlab8*", "cas_db")
c = db.cursor()

# 분별 날짜
date = np.genfromtxt('./classification_v2/date.csv', delimiter=',', dtype='str')

# csv 파일로 내보내기
w = open('classification_many_data.csv', 'w', encoding='utf-8')
wr = csv.writer(w)
```

```python
# 각 날짜별 데이터들 DB에서 가져오고 csv 파일로 저장
# 데이터는 cct, swr, uvb, uvi 순
for j in range(len(date)):
    sql = "select time(date), date(date), cct, cas_swr, uvb from natural_tracker left outer join cas_wave_ratio using(date) where date(date) = '"+ str(date[j]) + "' order by time(date)"
    c.execute(sql)
    rows = c.fetchall()

    sql = "select time(date), date(date), CAS_EryUVI from cas_uv_ratio where date(date) = '"+ str(date[j]) + "' order by time(date)"
    c.execute(sql)
    rows2 = c.fetchall()
    
    start = 0

#     for i in range(len(rows)):
#         if(str(rows[i][0]) == '6:00:00'):
#             print(rows[i][3])
    
    for i in range(len(rows)):
        if(str(rows[i][0]) == '6:00:00'):
            start = i
            break;
        if(str(rows[i][0]) == '6:01:00'):
            start = i
            break;

    for l in range(start, start+772):
        wr.writerow([rows[l][2], rows[l][3], rows[l][4]])
        
w.close()
db.close()
```

```python
# 데이터 받아오고 편차 계산히기
import_data = np.loadtxt('./classification_many_data.csv', delimiter=',')

one = import_data[:,0]
two = import_data[:,1]
thr = import_data[:,2]

cct = []
swr = []
uvb = []

delta_cct = []
delta_swr = []
delta_uvb = []

for cnt in range((int)(len(import_data) / 772)):
    temp = []
    temp2 = []
    temp3 = []
    for i in range(772):
        temp.append(one[i+772*cnt])
        temp2.append(two[i+772*cnt])
        temp3.append(thr[i+772*cnt])
    cct.append(temp)
    swr.append(temp2)
    uvb.append(temp3)
    
for cnt in range(88):
    temp = []
    temp2 = []
    temp3 = []
    for i in range(771):
        temp.append(cct[cnt][i+1] - cct[cnt][i])
        temp2.append(swr[cnt][i+1] - swr[cnt][i])
        temp3.append(uvb[cnt][i+1] - uvb[cnt][i])
    delta_cct.append(temp)
    delta_swr.append(temp2)
    delta_uvb.append(temp3)
    
print(np.shape(delta_swr))
```

```python
w = open('naturalLight_classification_criteria_data_cct.csv', 'w', encoding='utf-8')
wr = csv.writer(w)

row = ['date', '25', '50', '100', '150', '200', '250', '300', '350', '400', '450', '500', 'percent_1', 'percent_2', 'percent_3', 'percent_4', 'percent_5', 'percent_6', 'percent_7', 'percent_8', 'percent_9', 'percent_10', 'percent_11', 'hap', 'avg', 'grade']
wr.writerow(row)

for i in range(len(delta_cct)):
    # 편차의 절대값이 100 ~ 500인 경우의 갯수 저장
    up_dot_25 = 0
    up_dot_50 = 0
    up_dot_100 = 0
    up_dot_150 = 0
    up_dot_200 = 0
    up_dot_250 = 0
    up_dot_300 = 0
    up_dot_350 = 0
    up_dot_400 = 0
    up_dot_450 = 0
    up_dot_500 = 0

    # 위의 갯수가 하루 전체 데이터에서 차지하는 비율 저장
    percent_1 = 0
    percent_2 = 0
    percent_3 = 0
    percent_4 = 0
    percent_5 = 0
    percent_6 = 0
    percent_7 = 0
    percent_8 = 0
    percent_9 = 0
    percent_10 = 0
    percent_11 = 0

    # 편차의 합, 평균, 등급 저장
    hap = 0
    avg = 0
    grade = 0

    for j in range(len(delta_cct[i])):
        if(abs(delta_cct[i][j]) >= 25):
            up_dot_25 += 1
        if(abs(delta_cct[i][j]) >= 50):
            up_dot_50 += 1
        if(abs(delta_cct[i][j]) >= 100):
            up_dot_100 += 1
        if(abs(delta_cct[i][j]) >= 150):
            up_dot_150 += 1
        if(abs(delta_cct[i][j]) >= 200):
            up_dot_200 += 1
        if(abs(delta_cct[i][j]) >= 250):
            up_dot_250 += 1
        if(abs(delta_cct[i][j]) >= 300):
            up_dot_300 += 1
        if(abs(delta_cct[i][j]) >= 350):
            up_dot_350 += 1
        if(abs(delta_cct[i][j]) >= 400):
            up_dot_400 += 1
        if(abs(delta_cct[i][j]) >= 450):
            up_dot_450 += 1
        if(abs(delta_cct[i][j]) >= 500):
            up_dot_500 += 1
        
        hap += abs(delta_cct[i][j])
        avg = hap / 771

        percent_1 = round(up_dot_25 / 772 * 100, 2)
        percent_2 = round(up_dot_50 / 772 * 100, 2)
        percent_3 = round(up_dot_100 / 772 * 100, 2)
        percent_4 = round(up_dot_150 / 772 * 100, 2)
        percent_5 = round(up_dot_200 / 772 * 100, 2)
        percent_6 = round(up_dot_250 / 772 * 100, 2)
        percent_7 = round(up_dot_300 / 772 * 100, 2)
        percent_8 = round(up_dot_350 / 772 * 100, 2)
        percent_9 = round(up_dot_400 / 772 * 100, 2)
        percent_10 = round(up_dot_450 / 772 * 100, 2)
        percent_11 = round(up_dot_500 / 772 * 100, 2)

    if(percent_1 > 20.0):
        grade += 1
    
    if(hap > 18000.0):
        grade += 1
    
    if(percent_1 > 30.0):
        grade += 1
        
    if(hap > 30000.0):
        grade += 1

    wr.writerow([date[i], up_dot_25, up_dot_50, up_dot_100, up_dot_150, up_dot_200, up_dot_250, up_dot_300, up_dot_350, up_dot_400, up_dot_450, up_dot_500,percent_1, percent_2, percent_3, percent_4, percent_5, percent_6, percent_7, percent_8, percent_9,percent_10, percent_11, hap, avg, grade])
w.close();
```
