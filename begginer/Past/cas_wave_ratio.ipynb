```python
# 분별 날짜 파일에 있는 날짜 기준으로 DB에서 데이터 받아와서 저장하기

import MySQLdb
import os
import numpy as np
import csv

# MySQL DB 연결
db = MySQLdb.connect('210.102.142.13',"root", "witlab8*", "cas_db")
c = db.cursor()

# 분별 날짜
date = np.genfromtxt('./date.csv', delimiter=',', dtype='str')
# date = np.genfromtxt('./test_date.csv', delimiter=',', dtype='str')

print(len(date))

# csv 파일로 내보내기
w = open('cas_wave_ratio.csv', 'w', encoding='utf-8')
# w = open('cas_wave_ratio_test.csv', 'w', encoding='utf-8')
# w = open('rain_data.csv', 'w', encoding='utf-8')
wr = csv.writer(w)

# 각 날짜별 데이터들 DB에서 가져오고 csv 파일로 저장
# 데이터는 swr, mwr, lwr 순
for j in range(len(date)):
    sql = "select time(date), cas_swr, cas_mwr, cas_lwr from cas_wave_ratio where date(date) = '"+ str(date[j]) + "' order by date"
# sql = "select time(date), cas_swr, cas_mwr, cas_lwr from cas_wave_ratio where date(date) = '2017-08-17' order by date"
    c.execute(sql)
    rows = c.fetchall()

    start = 0

    for i in range(len(rows)):
        if(str(rows[i][0]) == '6:00:00'):
            start = i
            print(rows[start][0])
            break;
        if(str(rows[i][0]) == '6:01:00'):
            start = i
            print(rows[start][0])
            break;

    # start에 저장된 index부터 772개 데이터 가져와서 저장
    for l in range(start, start+772):
        wr.writerow([rows[l][1], rows[l][2], rows[l][3]])
    print(date[j] + " saved.")
# print("2017-08-17 saved.")
        
w.close()
db.close()
```

```python
# 데이터 받아오고 편차 계산히기
import_data = np.loadtxt('./cas_wave_ratio.csv', delimiter=',')
# import_data = np.loadtxt('./cas_wave_ratio_test.csv', delimiter=',')
# import_data = np.loadtxt('./rain_data.csv', delimiter=',')

w = open('swr_ratio.csv', 'w', encoding='utf-8')
# w = open('swr_ratio_test.csv', 'w', encoding='utf-8')
# w = open('rain_data_ratio.csv', 'w', encoding='utf-8')
wr = csv.writer(w)

print(len(date))

# swr = import_data[:,0] # swr
# mwr = import_data[:,1] # mwr
# lwr = import_data[:,2] # lwr

one = import_data[:,0] # swr
two = import_data[:,1] # mwr
thr = import_data[:,2] # lwr


# 날짜별 772개의 데이터 모두 저장, 2차원 배열
swr = []
mwr = []
lwr = []

# raw data들 저장
for cnt in range(len(date)): # 전체 데이터 날짜 수 (행)
    temp = []
    temp2 = []
    temp3 = []
    for i in range(772): # 한 날짜의 데이터들 한 행에 저장 (열)
        temp.append(one[i+772*cnt])
        temp2.append(two[i+772*cnt])
        temp3.append(thr[i+772*cnt])
    swr.append(temp)
    mwr.append(temp2)
    lwr.append(temp3)

ratio = []

for i in range(len(swr)):
    hap = 0
    swr_hap = 0
    temp = []
    for j in range(len(swr[i])):
        hap += swr[i][j] + mwr[i][j] + lwr[i][j]
        swr_hap += swr[i][j]
    temp.append(round(swr_hap / hap * 100, 2))
    ratio.append(temp)
for i in range(len(swr)):
    wr.writerow(ratio[i])

# hap = 0
# swr_hap = 0
# temp = []
# for j in range(772):
#     hap += swr[j] + mwr[j] + lwr[j]
#     swr_hap += swr[j]
# temp.append(round(swr_hap / hap * 100, 2))
# ratio.append(temp)
# wr.writerow(ratio[i])

w.close()
```
