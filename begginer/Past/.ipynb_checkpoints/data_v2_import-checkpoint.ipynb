```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager, rc

%matplotlib tk
```

```python
data = np.loadtxt('./v2data/v2/train1_v2_data.csv', delimiter=',')

bang = data[:, 0]
go = data[:, 1]
g = data[:, 2]
w = data[:, 3]
ptmt = data[:, 4]

bang0615 = []
bang0921 = []
bang0208 = []
bang0227 = []

go0615 = []
go0921 = []
go0208 = []
go0227 = []

g0615 = []
g0921 = []
g0208 = []
g0227 = []

w0615 = []
w0921 = []
w0208 = []
w0227 = []

ptmt0615 = []
ptmt0921 = []
ptmt0208 = []
ptmt0227 = []

for i in range(901):
    bang0615.append(bang[i])
    bang0921.append(bang[901 + i])
    bang0208.append(bang[901*2 + i])
    bang0227.append(bang[901*3 + i])
    
    go0615.append(go[i])
    go0921.append(go[901 + i])
    go0208.append(go[901*2 + i])
    go0227.append(go[901*3 + i])

    g0615.append(g[i])
    g0921.append(g[901 + i])
    g0208.append(g[901*2 + i])
    g0227.append(g[901*3 + i])

    w0615.append(w[i])
    w0921.append(w[901 + i])
    w0208.append(w[901*2 + i])
    w0227.append(w[901*3 + i])

    ptmt0615.append(ptmt[i])
    ptmt0921.append(ptmt[901 + i])
    ptmt0208.append(ptmt[901*2 + i])
    ptmt0227.append(ptmt[901*3 + i])
```
