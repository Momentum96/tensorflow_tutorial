```python
import tensorflow as tf

x_data = [[1, 2], [2, 3], [3, 1], [4, 3], [5,3], [6,2]]
y_data = [[0], [0], [0], [1], [1], [1]]

X = tf.placeholder(tf.float32, shape=[None, 2]) # n개의 2차원 입력
Y = tf.placeholder(tf.float32, shape=[None, 1]) # n개의 1차원 출력
W = tf.Variable(tf.random_normal([2, 1]), name = 'weight') # 2차원을 받아 1차원 결과
b = tf.Variable(tf.random_normal([1]), name='bias') # 결과와 같은 차원

H = tf.sigmoid(tf.matmul(X, W) + b) # 로지스틱 분류를 위한 가설 함수 설정(시그모이드) (0~1사이 값만 나옴)

cost = -tf.reduce_mean(Y*tf.log(H) + (1-Y) * tf.log(1-H)) # 시그모이드 함수의 비용함수

train = tf.train.GradientDescentOptimizer(learning_rate = 0.01).minimize(cost) # cost를 최소화하는 트레이닝

predicted = tf.cast(H > 0.5, dtype=tf.float32) # tf.cast 함수를 사용하여 H의 값이 0.5보다 크다면 True, 아니면 False
#dtype을 tf.float32로 하면 True는 1, False는 0으로 출력됨
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32)) # 정확도는 예측한 결과(1, 0)와 실제 Y값이 같은지 다른지(equal) 확인하고 평균을 냄
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001):
        cost_val, _ = sess.run([cost, train], feed_dict={X:x_data, Y:y_data})
        if step % 200 == 0:
            print(step, cost_val)
    
    h, c, y, a = sess.run([H, predicted, Y, accuracy], feed_dict={X:x_data, Y:y_data})
    print("\nHypothesis:", h, "\nCorrect (predicted) : ", c, "\nCorrect (Y): ", y, "\nAccuracy: ", a)
```
