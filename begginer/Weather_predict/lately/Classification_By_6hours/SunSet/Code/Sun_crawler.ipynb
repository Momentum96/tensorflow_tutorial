```python
import numpy as np
from selenium import webdriver
from bs4 import BeautifulSoup
import sys
import csv
import datetime

options = webdriver.ChromeOptions()
options.add_argument('headless')
options.add_argument('window-size=1920x1080')

driver = webdriver.Chrome('../../../../Past/chromedriver', chrome_options=options)
driver.implicitly_wait(3)

date = np.genfromtxt('../data/date.csv', delimiter=',', dtype='str')
# date = np.genfromtxt('../data/test_date.csv', delimiter=',', dtype='str')

w = open('../data/date_sunrise_sunset.csv', 'w', encoding='utf-8')
# w = open('../data/test_date_sunrise_sunset.csv', 'w', encoding='utf-8')
wr = csv.writer(w)

for i in range(len(date)):
    driver.get('https://astro.kasi.re.kr:444/life/pageView/9?date=' + date[i] + '&address=%EC%B6%A9%EB%82%A8+%EC%B2%9C%EC%95%88%EC%8B%9C+%EC%84%9C%EB%B6%81%EA%B5%AC+%EC%B2%9C%EC%95%88%EB%8C%80%EB%A1%9C+1223-24#')
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    notices = soup.select('p.tit > span.sunrise')
    notices2 = soup.select('p.tit > span.sunset')

    undata = []
    undata2 = []

    for n in notices:
        undata.append(n.text.strip())
    
    for n in notices2:
        undata2.append(n.text.strip())

    h = undata[0].split(' ')[0]
    m = undata[0].split(' ')[1]
    
    h2 = undata2[0].split(' ')[0]
    m2 = undata2[0].split(' ')[1]

    sunrise = h + ":" + m + ":00"
    sunrise = sunrise.replace("분","")
    sunrise = sunrise.replace("시","")
    sunrise = sunrise.replace("0", "", 1)
    
    sunset = h2 + ":" + m + ":00"
    sunset = sunset.replace("분","")
    sunset = sunset.replace("시","")
    
    print(sunrise, ", ", sunset)
    
    wr.writerow([date[i], sunrise, sunset])
w.close()
```
