```python
import pandas as pd
from pandas import DataFrame, Series
data = pd.read_csv('./coor_data.csv', names=[])
print(data.corr())
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20, 20))
sns.heatmap(data = data.corr(), annot=True, fmt='.2f', linewidths=.5, cmap='pink_r')
```
