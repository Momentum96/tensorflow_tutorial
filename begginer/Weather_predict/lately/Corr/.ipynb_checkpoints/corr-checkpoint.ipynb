```python
import MySQLdb
import os
import numpy as np
import csv
import datetime

db = MySQLdb.connect('210.102.142.13',"root", "witlab8*", "cas_db")
c = db.cursor()

# 분별 날짜
date = np.genfromtxt('./date_sunrise_sunset.csv', delimiter=',', dtype='str')

# csv 파일로 내보내기
w = open('./coordinate.csv', 'w', encoding='utf-8')
wr = csv.writer(w)

# 각 날짜별 데이터들 DB에서 가져오고 csv 파일로 저장
# 데이터는 cct, swr, uvb, uvi 순
for j in range(len(date)):
    sql = "select date(date), time(date), cct, cas_swr from natural_tracker left outer join cas_wave_ratio using(date) where date(date) = '"+ str(date[j][0]) + "' order by time(date)"
    c.execute(sql)
    rows = c.fetchall()
    
    # 일출 후 6시간과 일몰 후 6시간 데이터 사용, 일출 혹은 일몰 당시 데이터가 없을 경우 가장 가까운 다른 데이터로 변환
    sunrise = datetime.datetime.strptime(date[j][1], '%H:%M:%S')
    sunset = datetime.datetime.strptime(date[j][2], '%H:%M:%S')
    
    standard = datetime.datetime.strptime('1900-01-01 00:00:00', '%Y-%m-%d %H:%M:%S')
    
    sunrise = sunrise - standard
    sunset = sunset - standard
    
    start = 0
    end = 0
    
    for i in range(len(rows)):
        if(rows[i][1] >= sunrise):
            start = i
            break;
    
    for i in range(len(rows)):
        if(rows[i][1] >= sunset):
            end = i
            break;
    
    for i in range(start, end):
        wr.writerow([rows[i][0], rows[i][1], rows[i][2], rows[i][3]])
    print(date[j][0], " date saved.")
        
w.close()
db.close()
```
