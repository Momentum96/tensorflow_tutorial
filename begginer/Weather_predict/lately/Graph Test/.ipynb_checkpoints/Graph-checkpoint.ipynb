```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import MySQLdb
import os
import csv
import datetime
```

```python
db = MySQLdb.connect('210.102.142.13', "root", "witlab8*", "cas_db")
c = db.cursor()

date = '2018-02-08'

cct_data = []
swr_data = []

sql = "select time(date), date(date), cct, cas_swr from natural_tracker left outer join cas_wave_ratio using(date) where date(date) = '"+ date + "' order by time(date)"
c.execute(sql)
rows = c.fetchall()

start = 0
end = 0

for i in range(len(rows)):
    if(str(rows[i][0]) >= '7:00:00'):
        start = i
        print(rows[start][0], start)
        break;

end = start + 704
print(rows[end][0], end)
print(end - start)

for i in range(start, end):
    cct_data.append(rows[i][2])
    swr_data.append(rows[i][3])
    
delete = []

for i in range(len(swr_data)):
    if(not(swr_data[i] >=-10 and swr_data[i] <= 200)):
        delete.append(i)
    
for i in range(len(delete)):
    swr_data.pop(delete[i] - i)
```

```python
delta_cct = []
delta_cct2 = []

for i in range(len(cct_data) - 1):
    delta_cct.append(cct_data[i+1] - cct_data[i])
    delta_cct2.append(cct_data2[i+1] - cct_data2[i])

print(np.shape(delta_cct))
print(np.shape(delta_cct2))

print(max(cct_data))
print(min(cct_data))

print(max(cct_data2))
print(min(cct_data2))

print(max(swr_data))
print(min(swr_data))

print(max(swr_data2))
print(min(swr_data2))
```
